```python
!pip install fpdf2

from fpdf import FPDF
from fpdf.enums import XPos, YPos

class PDF(FPDF):
    def header(self):
        """Defines the header behavior for every page."""
        # Set font: Helvetica, Bold, size 12
        self.set_font("Helvetica", "B", 12)
        # Move to the right
        self.cell(80)
        # Title (Width, Height, Text, Border, Line break, Alignment)
        self.cell(30, 10, "Corporate Report", border=0, new_x=XPos.RIGHT, new_y=YPos.TOP, align="C")
        # Line break
        self.ln(20)

    def footer(self):
        """Defines the footer behavior for every page."""
        # Position at 1.5 cm from bottom
        self.set_y(-15)
        # Set font: Helvetica, Italic, size 8
        self.set_font("Helvetica", "I", 8)
        # Page number (using self.page_no() dynamically)
        self.cell(0, 10, f"Page {self.page_no()}/{{nb}}", border=0, new_x=XPos.RIGHT, new_y=YPos.TOP, align="C")

# 1. Initialize the PDF object (P = Portrait, mm = Millimeters, A4 size)
pdf = PDF(orientation="P", unit="mm", format="A4")

# 2. Set alias for total number of pages (used in footer as {nb})
pdf.alias_nb_pages()

# 3. Add a page
pdf.add_page()

# 4. Add Body Content
pdf.set_font("Helvetica", size=14)
pdf.cell(0, 10, "Welcome to the Automated PDF Generator!", new_x=XPos.LMARGIN, new_y=YPos.NEXT, align="L")
pdf.ln(5) # Add a small spacing line break

# Add a paragraph of multi-line text
pdf.set_font("Helvetica", size=11)
intro_text = (
    "This PDF was generated using Python and the fpdf2 library. "
    "It allows for dynamic header and footer generation, easy font styling, "
    "and precise grid-based multi-line text alignment."
)
pdf.multi_cell(0, 6, intro_text)
pdf.ln(10)

# 5. Create a Simple Data Table
pdf.set_font("Helvetica", "B", 11)
# Table Headers: Column width, Height, Title, Border
pdf.cell(40, 7, "ID", border=1)
pdf.cell(90, 7, "Item Name", border=1)
pdf.cell(40, 7, "Price", border=1, new_x=XPos.LMARGIN, new_y=YPos.NEXT)

# Table Data
pdf.set_font("Helvetica", size=11)
data = [
    ("001", "Widget A", "$10.00"),
    ("002", "Gadget B", "$25.50"),
    ("003", "Sub-assembly C", "$150.00"),
]

for row in data:
    pdf.cell(40, 7, row[0], border=1)
    pdf.cell(90, 7, row[1], border=1)
    pdf.cell(40, 7, row[2], border=1, new_x=XPos.LMARGIN, new_y=YPos.NEXT)

# 6. Output the file to disk
output_filename = "sample_document.pdf"
pdf.output(output_filename)

print(f"Success! PDF saved as '{output_filename}'")
```

```python
from fpdf import FPDF

class GeneralDeclarationPDF(FPDF):
    def header(self):
        # Setting up a thin border for the entire page layout
        self.set_line_width(0.3)
        self.rect(5, 5, 200, 287)

    def create_grid_row(self, heights, widths, texts, alignments, borders):
        """Helper function to safely align and print grid cells dynamically"""
        max_h = max(heights)
        start_x = self.get_x()
        start_y = self.get_y()

        for i, text in enumerate(texts):
            self.set_xy(start_x, start_y)
            # Use multi_cell to handle any internal line breaks inside cells cleanly
            self.multi_cell(widths[i], heights[i], text, border=borders[i], align=alignments[i])
            start_x += widths[i]

        self.set_xy(5, start_y + max_h)

# Initialize PDF
pdf = GeneralDeclarationPDF(orientation="P", unit="mm", format="A4")
pdf.set_auto_page_break(auto=True, margin=5)
pdf.add_page()
pdf.set_font("Helvetica", size=9)

# --- HEADER SECTION (Metadata Grid) ---
# Row 1: Owner & Marks
pdf.create_grid_row([6, 6], [100, 100], ["Owner or Operator: SPICEJET", "Marks of Nationality and Registration:"], ["L", "L"], [1, 1])

# Row 2: Flight No & Departure
pdf.create_grid_row([6, 6], [100, 100], ["Flight No: SG5023", "Departure: DUBAI"], ["L", "L"], [1, 1])

# Row 3: Routing & Title
pdf.create_grid_row([14, 14], [100, 100], ["Routing: BOM\\DXB\\BOM", "GENERAL DECLARATION\nICAO Annex 9 Appendix 1\n(Outward / Inward)"], ["L", "C"], [1, 1])

# Row 4: Dates
pdf.create_grid_row([6, 6], [100, 100], ["Date: 03/06/2026 14:26", "Date: 03/06/2026 11:10"], ["L", "L"], [1, 1])

pdf.ln(2)

# --- CREW & PASSENGER SPLIT SECTION ---
# Custom split column mapping to handle the side-by-side tables
start_y = pdf.get_y()

# Left Column: Crew Details Table
pdf.set_xy(5, start_y)
pdf.set_font("Helvetica", "B", 8)
crew_headers = ["Crew Name", "Desig", "Nationality", "Sno.", "Gen", "DOB"]
crew_widths = [35, 15, 22, 8, 8, 12]

for i, h in enumerate(crew_headers):
    pdf.cell(crew_widths[i], 6, h, border=1, align="C")
pdf.ln()

pdf.set_font("Helvetica", size=8)
crew_data = [
    ["CAPT ISLAM", "CAPT", "PAKISTANI", "1", "M", ""],
    ["FO MUHAMMAD SOHAIL ASGHAR", "FO", "BRAZILIAN", "2", "", ""],
    ["ZARAQ KHAN", "CC", "PAKISTANI", "3", "", ""]
]

for row in crew_data:
    current_x = 5
    # Calculate row height dynamically based on the longest string (Name)
    pdf.set_xy(current_x, pdf.get_y())
    cell_lh = 4 if len(row[0]) > 20 else 6

    pdf.multi_cell(crew_widths[0], cell_lh, row[0], border=1, align="L")
    current_x += crew_widths[0]

    for i in range(1, 6):
        pdf.set_xy(current_x, pdf.get_y() - (cell_lh if len(row[0]) > 20 else 6))
        pdf.cell(crew_widths[i], cell_lh, row[i], border=1, align="C")
        current_x += crew_widths[i]
    pdf.ln()

# Right Column: Passenger Manifest Info
end_crew_y = pdf.get_y()
pdf.set_xy(105, start_y)
pdf.set_font("Helvetica", size=7.5)

passenger_info = (
    "Number of passengers on this stage(2)\n\n"
    "Departure Place:\n\n"
    "Embarking:\n"
    "Through on same flight:\n\n"
    "Arrival place:\n\n"
    "Disembarking:\n"
    "Through on same flight:\n\n"
    "(1) To be completed only when required by the state\n"
    "(2) Not to be completed when passenger manifest are presented "
    "and to be completed only when required by the state"
)
pdf.multi_cell(100, 3.8, passenger_info, border=1, align="L")
end_pass_y = pdf.get_y()

# Synchronize Y position after the split layout
max_split_y = max(end_crew_y, end_pass_y)
pdf.set_xy(5, max_split_y + 2)

# --- HEALTH DECLARATION SECTION ---
pdf.set_font("Helvetica", "B", 9)
pdf.cell(130, 5, "DECLARATION OF HEALTH", border="TLR", ln=1)
pdf.set_font("Helvetica", size=7.5)

health_text = (
    "Persons on board known to be suffering from illness other than airsickness or the effects of accidents, "
    "as well as those cases of illness disembarked during the flight - other condition on board which may "
    "lead to spread of disease:\n\n"
    "Details of each disinsecting or sanitary treatment (place, date, time, method) during the flight. If no "
    "disinsecting has been carried out during the flight give details of most recent disinsecting."
)

# Use dynamic dual column for Health text + Place definition column
health_y_start = pdf.get_y()
pdf.multi_cell(130, 3.5, health_text, border="BLR", align="L")
health_y_end = pdf.get_y()

pdf.set_xy(135, health_y_start - 5)
place_text = '(3) "PLACE" Column always to list origin, every enroute stop and destination.\n\n\n\nFOR OFFICIAL USE ONLY'
pdf.multi_cell(70, 4.3, place_text, border=1, align="L")
place_y_end = pdf.get_y()

# Synchronize Y position again
max_health_y = max(health_y_end, place_y_end)
pdf.set_xy(5, max_health_y + 4)

# --- SIGNATURE BLOCK ---
sig_text = (
    "I declare that all statements and particulars contained in this General Declaration, and in any "
    "supplementary forms required to be presented with this General Declaration are complete, exact and "
    "true to the best of my knowledge and that all through passengers will continue/have continued on the flight."
)
pdf.multi_cell(200, 4, sig_text, border=0, align="L")
pdf.ln(12)

# Signature line placeholders
pdf.set_font("Helvetica", size=8)
pdf.create_grid_row(
    [10, 10],
    [100, 100],
    ["\n_____________________________________\nSgd. if required Crew member concerned.", "\n_____________________________________\nSIGNATURE Authorised Agent or Pilot-In-Command"],
    ["C", "C"],
    [0, 0]
)

# --- FOOTER ---
pdf.set_xy(5, 282)
pdf.set_font("Helvetica", size=8)
pdf.cell(200, 5, "Page 1", border=0, align="C")

# Save PDF output
pdf.output("general_declaration.pdf")
```

```python
!pip install fpdf2
from fpdf import FPDF
from datetime import datetime

def get_current_datetime():
    return datetime.now().strftime("%d/%m/%Y %H:%M")

class PDF(FPDF):
    def header(self):
        self.set_font("Arial", "B", 14)
        self.cell(50, 8, "GENERAL DECLARATION", ln=False, align="C")
         # 2. Date and Time Subheading
        self.set_font("Arial", "", 10)  # Regular font, smaller size
        self.cell(0, 8, "Date: "+get_current_datetime(), ln=True, align="R")

        # 3. Add a small vertical spacing line break
        self.ln(2)

        self.set_font("Arial", "B", 10)
        self.cell(0, 6, "ICAO Annex 9 Appendix 1", ln=True, align="C")
        self.cell(0, 6, "(Outward / Inward)", ln=True, align="C")
        self.ln(5)

    def section_title(self, title):
        self.set_font("Arial", "B", 11)
        self.cell(0, 8, title, ln=True)
        self.set_font("Arial", "", 10)


pdf = PDF()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()

# ------------------------------------------------------------------
# Flight Information
# ------------------------------------------------------------------
pdf.set_font("Arial", "B", 10)
pdf.cell(100, 6, "Owner or Operator : SPICEJET", 0, 0)
pdf.set_font("Arial", "", 10)
pdf.cell(0, 6, "Date : 03/06/2026 14:26", ln=True)

pdf.cell(100, 6, "Flight No : SG5023", 0, 0)
pdf.cell(0, 6, "Flight Date : 03/06/2026 11:10", ln=True)

pdf.cell(100, 6, "Departure : DUBAI", 0, 0)
pdf.cell(0, 6, "Routing : BOM\\DXB\\BOM", ln=True)

pdf.cell(
    0,
    6,
    "Marks of Nationality and Registration : _______________________",
    ln=True,
)

pdf.ln(4)

# ------------------------------------------------------------------
# Passenger Information
# ------------------------------------------------------------------
pdf.section_title("Passenger Information")

pdf.cell(95, 6, "Departure Place :", border=1)
pdf.cell(95, 6, "Arrival Place :", border=1, ln=True)

pdf.cell(95, 6, "Embarking :", border=1)
pdf.cell(95, 6, "Disembarking :", border=1, ln=True)

pdf.cell(95, 6, "Through on same flight :", border=1)
pdf.cell(95, 6, "Through on same flight :", border=1, ln=True)

pdf.cell(0, 6, "Number of passengers on this stage (2): __________", ln=True)

pdf.ln(4)

# ------------------------------------------------------------------
# Crew Table
# ------------------------------------------------------------------
pdf.section_title("Crew Details")

headers = [
    ("SNo", 12),
    ("Crew Name", 60),
    ("Desig", 20),
    ("Nationality", 40),
    ("Gender", 18),
    ("DOB", 35),
]

pdf.set_font("Arial", "B", 9)
for h, w in headers:
    pdf.cell(w, 8, h, border=1, align="C")
pdf.ln()

pdf.set_font("Arial", "", 9)

crew_data = [
    ["1", "CAPT ISLAM", "CAPT", "PAKISTANI", "M", ""],
    ["2", "MUHAMMAD SOHAIL ASGHAR", "FO", "BRAZILIAN", "M", ""],
    ["3", "ZARAQ KHAN", "CC", "PAKISTANI", "M", ""],
]

for row in crew_data:
    pdf.cell(12, 8, row[0], border=1)
    pdf.cell(60, 8, row[1], border=1)
    pdf.cell(20, 8, row[2], border=1)
    pdf.cell(40, 8, row[3], border=1)
    pdf.cell(18, 8, row[4], border=1)
    pdf.cell(35, 8, row[5], border=1)
    pdf.ln()

pdf.ln(5)

# ------------------------------------------------------------------
# Declaration of Health
# ------------------------------------------------------------------
pdf.section_title("DECLARATION OF HEALTH")

health_text = """
Persons on board known to be suffering from illness other than airsickness
or the effects of accidents, as well as those cases of illness disembarked
during the flight:

______________________________________________________________

Any other condition on board which may lead to spread of disease:

______________________________________________________________

Details of each disinsecting or sanitary treatment (place, date, time,
method) during the flight. If no disinsecting has been carried out during
the flight, give details of the most recent disinsecting.

______________________________________________________________

Sgd. if required ______________________________________
Crew member concerned.
"""

pdf.multi_cell(0, 6, health_text)

pdf.ln(2)

# ------------------------------------------------------------------
# General Declaration
# ------------------------------------------------------------------
declaration = """
I declare that all statements and particulars contained in this General
Declaration, and in any supplementary forms required to be presented with
this General Declaration are complete, exact and true to the best of my
knowledge and that all through passengers will continue/have continued on
the flight.
"""

pdf.multi_cell(0, 6, declaration)

pdf.ln(8)

pdf.cell(
    0,
    6,
    "SIGNATURE _____________________________________",
    ln=True,
)

pdf.cell(
    0,
    6,
    "Authorised Agent or Pilot-In-Command",
    ln=True,
)

pdf.ln(5)

# ------------------------------------------------------------------
# Notes
# ------------------------------------------------------------------
pdf.section_title("Notes")

notes = """
(1) To be completed only when required by the State.

(2) Not to be completed when passenger manifests are presented and to be
completed only when required by the State.

(3) 'PLACE' column always to list origin, every enroute stop and destination.

FOR OFFICIAL USE ONLY
"""

pdf.multi_cell(0, 5, notes)

# Save PDF
pdf.output("general_declaration.pdf")

print("PDF generated successfully: general_declaration.pdf")
```
